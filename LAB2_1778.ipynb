```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
hs = fetch_california_housing()
hs
```

```python
x = hs.data
y = hs.target
x
```

```python
y
```

```python
columns = hs.feature_names
```

```python
print(hs.DESCR)

```

```python
# Identify the type of "x"
print(type(x))
```

```python
# Create a data frame
hs_df = pd.DataFrame(hs.data)
hs_df.columns = columns
print(hs_df.head())

```

```python
# Draw a box plot for the feature "HouseAge" and print the IQR for each column
plt.boxplot(hs_df['HouseAge'])
plt.title('Box plot for HouseAge')
plt.show()

for col in hs_df.columns:
    q1 = hs_df[col].quantile(0.25)
    q3 = hs_df[col].quantile(0.75)
    iqr = q3 - q1
    print(f"IQR for {col}: {iqr}")

```

```python
# Draw a plot to visualize the correlation between each feature
corr_matrix = hs_df.corr()
plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.show()
```

```python

# Apply a histogram plot and find the inference out of it for the feature (PRICE)
plt.hist(y, bins=30)
plt.xlabel('House Price')
plt.ylabel('Frequency')
plt.title('Histogram of House Prices')
plt.show()

# The histogram shows that most of the house prices are between 10-25, with a few outliers beyond 40.

```

```python
# Draw a scatter plot for the Population Vs. value of house
plt.scatter(hs_df['HouseAge'], y)
plt.xlabel('Population')
plt.ylabel('HouseAge')
plt.title('Scatter Plot of Population Vs. HouseAge')
plt.show()
```
