```python
import pandas as pd
import numpy as np

# Task 1
data = np.random.rand(4, 4)
```

```python
data
```

```python
df = pd.DataFrame(data, columns=[' 1', ' 2', ' 3', ' 4'])

df
```

```python

# Task 2
df.rename(columns={' 1': 'Random value 1',
                   ' 2': 'Random value 2',
                   ' 3': 'Random value 3',
                   ' 4': 'Random value 4'}, inplace=True)

df
```

```python
# Task 3
descriptive_stats = df.describe()
descriptive_stats
```

```python
# Task 4
null_values = df.isnull().sum()
data_types = df.dtypes

null_values
```

```python
data_types
```

```python
# Task 5
random_value_2_loc = df.loc[:, 'Random value 2']
random_value_3_loc = df.loc[:, 'Random value 3']

random_value_2_idx = df.iloc[:, 1]
random_value_3_idx = df.iloc[:, 2]

# Display results
print("Task 1: DataFrame 'df'")
print(df)

print("\nTask 2: DataFrame after column renaming")
print(df)

print("\nTask 3: Descriptive Statistics")
print(descriptive_stats)

print("\nTask 4: Null Values and Data Types")
print("Null Values:")
print(null_values)
print("\nData Types:")
print(data_types)

print("\nTask 5: Random value 2 and Random value 3 columns")
print("Using .loc:")
print(random_value_2_loc)
print(random_value_3_loc)
print("\nUsing iloc:")
print(random_value_2_idx)
print(random_value_3_idx)
```
